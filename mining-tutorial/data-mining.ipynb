### We are working with a dataset that contains 21,613 observations (count id), mean price is approximately 540k, median price is approximately 450k, and average house's area is 2080 ft2.

```python
fig = plt.figure(figsize=(12,6))
sqft = fig.add_subplot(121)
cost = fig.add_subplot(122)

sqft.hist(df.sqft_living, bins=80)
sqft.set_xlabel('Ft^2')
sqft.set_title("Histogram of House Square Footage")

cost.hist(df.price, bins=80)
cost.set_xlabel('Price ($)')
cost.set_title("Histogram of Housing Prices")

plt.show()
```

### Using matplotlib (plt) we printed two histograms to observe the distribution of housing prices and square footage. Left histogram plots number of houses on y, with ft^2 on x. Right histogram plots number of houses on y, with price on x.

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
# doing regression analysis, we need statsmodels
# to get the least squares regression estimator functions;
# the 'ordinary least squares' module

# linear regression with OLS and only 2 vars
# Reg = ols('Dependent variable ~ independent variable(s)'', dataframe).fit())
# print(Reg.summary())
m = ols('price ~ sqft_living', df).fit()
print(m.summary())
```

### The extremely high t-value of 144.920 and P>|t| of 0% means this relationship has a near zero chance of being due to statistical variation or chance. Every additional 100 square feet above the average, the house price would rise $28,000.

```python
m = ols('price ~ sqft_living + bedrooms + grade + condition', df).fit()
print(m.summary())
```

### The multivariate regression output implies that by using additional independent variables, such as the number of bedrooms, we can provide a model that fits the data better, as the R-squared regression has increased to 0.555. This means that we went from being able to explain about 49.3% of the variation in the model to 55.5% with the addition of a few more independent variables.

```python
sns.jointplot(x="sqft_living", y="price", data=df, kind="reg", fit_reg=True, size=7)
plt.show()
```
